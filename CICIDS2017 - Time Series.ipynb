```python
from sklearn.preprocessing import MinMaxScaler
# normalizing the 'Flow Duration' column we can use to avoid overflow (floating point precision issues) 
scaler = MinMaxScaler()
df['Flow Duration Normalized'] = scaler.fit_transform(df[['Flow Duration']])
plt.plot(df['Flow Duration Normalized'][::1000], color='blue')
df['Flow Duration Normalized'].head()
```

```python
from sklearn.preprocessing import StandardScaler # - 3 to 3 (if normally distributed)

scaler = StandardScaler()
df['Flow Duration Normalized'] = scaler.fit_transform(df[['Flow Duration']])
plt.plot(df['Flow Duration Normalized'][::1000], color='red')
df['Flow Duration Normalized'].head()
```

We're going to use MinMax() normalization because it will give us a bounded results, which will help us better if we're using it with a quantum circuit. 

```python
import numpy as np

# Convert normalized data [0,1] to angles [-π, +π]
df['Flow Duration Angles'] = (df['Flow Duration Normalized'] * 2 * np.pi) - np.pi

# Verify the encoding
print(f"Angle range: [{df['Flow Duration Angles'].min():.3f}, {df['Flow Duration Angles'].max():.3f}]")
print(f"Sample angles: {df['Flow Duration Angles'].head()}")
```

```python
# Plot the angle-encoded data
plt.figure(figsize=(12, 4))
plt.plot(df['Flow Duration Angles'][::20000], color='red', alpha=0.7)
plt.title('Flow Duration - Angle Encoded [-π, +π]')
plt.ylabel('Angle (radians)')
plt.xlabel('Sample Index')
plt.axhline(y=0, color='black', linestyle='--', alpha=0.3)
plt.axhline(y=np.pi, color='black', linestyle='--', alpha=0.3)
plt.axhline(y=-np.pi, color='black', linestyle='--', alpha=0.3)
plt.grid(True, alpha=0.3)
plt.show()
```
